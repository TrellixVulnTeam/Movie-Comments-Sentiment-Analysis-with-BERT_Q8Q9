```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('/content/drive/MyDrive/bert_data.csv',encoding='utf-8')
```

```
df2 = {'movie_name': 'else', 'predict_rate':3.7693537506584938}

df = df.append(df2, ignore_index = True)
```

```
df
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
cos_list=[]
for i in range(len(df)):
  cos_list.append(df['predict_rate'][i])
#print('=========')
#print(final)
cos_list = np.array(cos_list).reshape(-1, 1)
scaler = MinMaxScaler()
cos_list = scaler.fit_transform(cos_list)
movie_content_dict = dict([str(df["movie_name"][i]), cos_list[i].item() * 20] for i in range(len(cos_list)))
movie_content_dict
```

```
movie_list=list(movie_content_dict.items())
with open('bert_minmax.csv','w', newline='') as csvfile:

  # 讀取 CSV 檔案內容
  writer = csv.writer(csvfile)

  # 以迴圈輸出每一列
  for i in range(len(movie_list)):
    writer.writerow([movie_list[i][0],movie_list[i][1]])
```
